Ahora interactuemos un poco con datos. Para eso, vamos a utilizar pandas para levanar una serie de datos que se conoce como el Cuarteto de Anscombe (que vamos a usar la clase que viene!). Pandas posee una gran versatilidad. Permite leer archivos de distintas clases, txt, csv, json.
En los datos vemos que hay cuatro series, cada una compuesta por 11 puntos y el valor de una funci√≥n para cada uno de ellos.

Veamos tambi√©n como filtrar estos archivos, qued√°ndonos solo con la primera serie.

```python
anscombe = pd.read_json('/content/sample_data/anscombe.json')

print(anscombe)

x = anscombe[anscombe['Series']=='I']['X']
y = anscombe[anscombe['Series']=='I']['Y']

plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
series_list = anscombe['Series'].unique()

# Creamos una cuadr√≠cula 2x2 para los gr√°ficos de los 4 sets de datos
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# De esta manera pasamos de 2 componentes a 1 para utilizar un solo √≠ndice para referirnos a cada subplot
axs = axs.flatten()

# Iteramos para los cuatro valores de la serie y plotteamos cada uno
for i, series in enumerate(series_list):
    print(f'Estoy trabajando con la serie {series} e i= {i}')
    data = anscombe[anscombe['Series'] == series]
    axs[i].scatter(data['X'], data['Y'])
    axs[i].set_title(series)
    axs[i].set_xlabel('X')
    axs[i].set_ylabel('Y')
plt.tight_layout()
```

Y finalmente, ajustemos los datos para las cuatro curvas. En este caso, vamos a usar la herramienta m√°s conocida que es el m√©todo _curve-fit_ de _scipy.optimize_ aunque es un m√©todo que no recomiendo pero eso se los cuento la clase que viene. Adem√°s vamos a sacar algunos estad√≠sticos sobre este set de datos.

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axs = axs.flatten()
mean = []
sd = []
a = []
b = []

def linear_func(x, a, b):
    return a * x + b

# Iteramos para los cuatro valores de la serie y plotteamos cada uno
for i, series in enumerate(series_list):
    data = anscombe[anscombe['Series'] == series]
    mean_i = np.mean(data['Y'])
    sd_i = np.std(data['Y'])
    mean.append(mean_i)
    sd.append(sd_i)
    popt, pcov = scp.curve_fit(linear_func, data['X'], data['Y'])

    a.append(popt[0])
    b.append(popt[1])

    x_fit = np.linspace(min(data['X']), max(data['X']), 100)
    y_fit = linear_func(x_fit, *popt)
    axs[i].plot(x_fit, y_fit, 'r-', label='Linear Fit')

    axs[i].scatter(data['X'], data['Y'])
    axs[i].set_title(series)
    axs[i].set_xlabel('X')
    axs[i].set_ylabel('Y')
plt.tight_layout()
plt.show()

results = pd.DataFrame({
    'Series': series_list,
    'Mean': mean,
    'Standard Deviation': sd,
    'a': a,
    'b': b
})

results.to_csv('resultados.csv')
```
