También repasemos los tres loops comunes : _for_, _while_ e _if_.

En el caso del _for_ la cantidad de iteraciones está definida a priori. En cambio en el _while_ se va a repetir el loop mientras la condición sea verdadera.  

```python
for i in range(1,10):
  print(cuadrado(i))

if (1==0):
  print("Es verdad")
else:
  print("Es mentira")

cont = 0
while cont < 10:
  cont += 1
  if cont != 10:
    print(f"El contador es {cont}, sigo hasta 10")
  else:
    print("Terminé")


```

Ahora interactuemos un poco con datos. Para eso, vamos a utilizar pandas para levanar una serie de datos que se conoce como el Cuarteto de Anscombe (que vamos a usar la clase que viene!). Pandas posee una gran versatilidad. Permite leer archivos de distintas clases, txt, csv, json.
En los datos vemos que hay cuatro series, cada una compuesta por 11 puntos y el valor de una función para cada uno de ellos.

Veamos también como filtrar estos archivos, quedándonos solo con la primera serie.

```python
anscombe = pd.read_json('/content/sample_data/anscombe.json')

print(anscombe)

x = anscombe[anscombe['Series']=='I']['X']
y = anscombe[anscombe['Series']=='I']['Y']

plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
series_list = anscombe['Series'].unique()

# Creamos una cuadrícula 2x2 para los gráficos de los 4 sets de datos
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# De esta manera pasamos de 2 componentes a 1 para utilizar un solo índice para referirnos a cada subplot
axs = axs.flatten()

# Iteramos para los cuatro valores de la serie y plotteamos cada uno
for i, series in enumerate(series_list):
    print(f'Estoy trabajando con la serie {series} e i= {i}')
    data = anscombe[anscombe['Series'] == series]
    axs[i].scatter(data['X'], data['Y'])
    axs[i].set_title(series)
    axs[i].set_xlabel('X')
    axs[i].set_ylabel('Y')
plt.tight_layout()
```

Y finalmente, ajustemos los datos para las cuatro curvas. En este caso, vamos a usar la herramienta más conocida que es el método _curve-fit_ de _scipy.optimize_ aunque es un método que no recomiendo pero eso se los cuento la clase que viene. Además vamos a sacar algunos estadísticos sobre este set de datos.

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axs = axs.flatten()
mean = []
sd = []
a = []
b = []

def linear_func(x, a, b):
    return a * x + b

# Iteramos para los cuatro valores de la serie y plotteamos cada uno
for i, series in enumerate(series_list):
    data = anscombe[anscombe['Series'] == series]
    mean_i = np.mean(data['Y'])
    sd_i = np.std(data['Y'])
    mean.append(mean_i)
    sd.append(sd_i)
    popt, pcov = scp.curve_fit(linear_func, data['X'], data['Y'])

    a.append(popt[0])
    b.append(popt[1])

    x_fit = np.linspace(min(data['X']), max(data['X']), 100)
    y_fit = linear_func(x_fit, *popt)
    axs[i].plot(x_fit, y_fit, 'r-', label='Linear Fit')

    axs[i].scatter(data['X'], data['Y'])
    axs[i].set_title(series)
    axs[i].set_xlabel('X')
    axs[i].set_ylabel('Y')
plt.tight_layout()
plt.show()

results = pd.DataFrame({
    'Series': series_list,
    'Mean': mean,
    'Standard Deviation': sd,
    'a': a,
    'b': b
})

results.to_csv('resultados.csv')
```
